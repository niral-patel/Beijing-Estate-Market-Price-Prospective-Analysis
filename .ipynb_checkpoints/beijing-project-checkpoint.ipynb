```python
dfColumnNames = ['price','totalPrice']
print(dfColumnNames)
# #for j in df2ColumnNames:
for j in dfColumnNames:
    try:
        print (j)
        quartile = df[j]
        q1, q3 = np.percentile(quartile, 25), np.percentile(quartile, 75)
        iqr = q3 - q1
        #now calculate outlier cut_off by using iqr
        cut_off_iqr = iqr * 1.5
        print (f'interquartile range for the given data-set is {cut_off_iqr}')
        #Shape of the dataframe before outlier removal
        shape_before_outlier_eliminate = df[j].shape
        print(f"# of rows and columns  in the dataframe before the elimination of Outlier {shape_before_outlier_eliminate}")
        #By using IQR
        upper_lim_iqr, lower_lim_iqr = q3 + cut_off_iqr, q1 - cut_off_iqr
        print (f'''Outliers based on the IQR method:
        upper limit of the Box Plot is {upper_lim_iqr}
        lower limit of the Box Plot is {lower_lim_iqr}''')
        #After finding out the outliers by using IQR method, streamlining our data to fit within these boundaries eliminating the outliers.
        df = df[df[j] <= upper_lim_iqr] # Remove from data set that is above the upper threshold value
        df = df[df[j] >= lower_lim_iqr] # Remove from data set that is below the lower threshold value
        last_value = df[j].shape #shape after removing the outliers by using IPQ method 
        print (last_value)
    except: 
           print("An error occured")
```

#### Split the data set X and y into X_train, y_train which will be used for training the model
#### X_test and y_test will be used for validating and testing the efficiency of the trained model
#### The chosen ratio of division for the data is 80: 20 for train: test data set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=73)
```

#### Our chosen approach for model is Regression. Thus experimenting with variety of different regression models to observe its performance on test data post training cycle.

```python
from dmba import backward_elimination
from dmba import AIC_score #, BIC_score
def train_model(variables):

    model = LinearRegression()

    model.fit(X_train[variables], y_train)

    return model

 

def score_model(model, variables):

    return AIC_score(y_train, model.predict(X_train[variables]), model)

 

best_model, best_variables = backward_elimination(X_train.columns, train_model, score_model, verbose=True)

print(best_variables)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

# Fit a lasso model on the training set
model = Lasso().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Real Estate pricing')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='magenta')
plt.show()

#Comparison of the predicted values of model and the actual test values in y_test
predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted labels: ', np.round(predictions)[:10])
print('Actual labels   : ' ,y_test[:10])
```

# Comparing the accuracies of various Regression models

```python
from sklearn.metrics import accuracy_score
Regressions = [
    Lasso(),
    BayesianRidge(),
    LinearRegression(),
    RandomForestRegressor()]

ModelAccuracy=[]
RegModel=[]
for regression in Regressions:
    fit = regression.fit(X_train ,y_train)
    score_train = fit.score(X_test, y_test)
    score_test = fit.score(X_train, y_train)
    pred=fit.predict(X_test)
    ModelAccuracy.append(score_train)
    RegModel.append(regression.__class__.__name__)
    print("Accuracy of "+regression.__class__.__name__+" is" )
    print( "Training score: ", score_train)
    print( "Testing score: ",score_test)
    print('\n')
```

```python
Index = [1,2,3,4]
plt.bar(Index,ModelAccuracy)
plt.xticks(Index, RegModel,rotation=45)
plt.ylabel('Model Accuracy')
plt.xlabel('Regression Model')
plt.title('Accuracies of regression Models')
```
